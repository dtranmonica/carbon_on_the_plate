**STAGE 1: TOTAL EMISSIONS RANKINGS**  

The analysis identifies food products with the highest and the lowest total emissions. In the further step, results will be visualised in bar plot with explanations.

```python
#Importing necessary libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

#Importing dataset
food = pd.read_csv('clean.csv')
food.head(10)
```

```python
#Finding the top ten food products with the highest emissions
highest_emitters = food.sort_values(by = 'total_emissions', ascending = False)
highest_emitters.head(10)
```

```python
#Finding the top ten food products with the lowest emissions
lowest_emitters = food.sort_values(by = 'total_emissions', ascending = True)
lowest_emitters.head(10)
```
